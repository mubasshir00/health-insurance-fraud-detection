```python
import pandas as pd
```

```python
# Load both results
rf_df = pd.read_csv("/home/mubasshir/Desktop/Research/Insurance/health-insurance-fraud-detection/RF_res.csv")
xgb_df = pd.read_csv("/home/mubasshir/Desktop/Research/Insurance/health-insurance-fraud-detection/XGB.csv")
```

```python
# Merge on Provider
combined_df = pd.merge(rf_df, xgb_df, on='Provider', how='inner')
```

```python
# Combined weighted probability
weight_rf = 0.4
weight_xgb = 0.6
```

```python
combined_df['FraudProbability_Combined'] = (
    combined_df['FraudProbability_RF'] * weight_rf +
    combined_df['FraudProbability_XGB'] * weight_xgb
)
```

```python

# Final fraud prediction based on threshold
threshold = 0.5
combined_df['PredictedFraud_Combined'] = (
    combined_df['FraudProbability_Combined'] >= threshold
).astype(int)
```

```python
output_path= "fraud_prediction_RF_XGB_combined.csv"
```

```python
combined_df[
    ['Provider',
     'PredictedFraud_RF',
     'FraudProbability_RF',
     'PredictedFraud_XGB',
     'FraudProbability_XGB',
     'FraudProbability_Combined',
     'PredictedFraud_Combined']
].to_csv(output_path, index=False)
```
